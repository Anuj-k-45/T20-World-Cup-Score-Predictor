```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pickle.load(open("dataset_level2.pkl", "rb"))
```

```python
# batting team
# bowling team
# city
# current score
# balls left
# wickets left
# current rr
# last five overs runs
```

```python
df.isnull().sum()
```

```python
df[df["city"].isnull()]["venue"].value_counts()
```

```python
cities = np.where(df["city"].isnull(), df["venue"].str.split().apply(lambda x : x[0]), df["city"])
```

```python
df["city"] = cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["venue"], inplace=True)
```

```python
df
```

```python
eligible_city = df["city"].value_counts()[df["city"].value_counts() > 600].index.tolist()
```

```python
df = df[df["city"].isin(eligible_city)]
```

```python
df
```

```python
df.info()
```

```python
df.loc[:, 'runs'] = pd.to_numeric(df['runs'], errors='coerce')     #incase if any data is not numeric
```

```python
grouped = df.groupby('match_id')["runs"]

for group_name, group_data in grouped:
    print(f"Group: {group_name}")
    print(group_data.head()) 
```

```python
df["current_score"] = df.groupby("match_id")['runs'].cumsum()
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=="0" else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df["crr"] = (df["current_score"]*6)/df["balls_bowled"]
```

```python
df
```

```python
match_ids = df['match_id'].unique()
groups = df.groupby("match_id")

last_five = []

# Iterate over each match and calculate the rolling sum
for id in match_ids:
    # Get the group corresponding to the match
    group = groups.get_group(id)
    
    # Calculate the rolling sum for the 'runs' column with a window of 30
    rolling_sums = group['runs'].rolling(window=30).sum()
    
    # Extend the 'last_five' list with the values of the rolling sum
    last_five.extend(rolling_sums.values.tolist())

last_five
```

```python
df["last_five"] = last_five
```

```python
df
```

```python
final_df = df.groupby("match_id")["runs"].sum().reset_index().merge(df, on="match_id")
```

```python
final_df
```

```python
final_df = final_df[["batting_team", "bowling_team", "city", "current_score", "balls_left", "wickets_left", "crr", "last_five", "runs_x"]]
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

Model Training

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
